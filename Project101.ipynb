```python
import pandas as pd 
```

```python
df=pd.read_csv(r"C:\Users\LENOVO\Downloads\student-scores.csv")
df
```

```python
df=df.drop_duplicates()
df
```

```python
df.fillna({'score': 0}, inplace=True)  # Example: Fill missing scores with 0
df.dropna(subset=['id'], inplace=True)  # Remove rows with missing student IDs
df
```

```python
#converting datatypes : id->integer scores->numeric
df['id'] = pd.to_numeric(df['id'], errors='coerce')
df['math_score'] = pd.to_numeric(df['math_score'], errors='coerce')
df['history_score'] = pd.to_numeric(df['history_score'], errors='coerce')
df['physics_score'] = pd.to_numeric(df['physics_score'], errors='coerce')
df['chemistry_score'] = pd.to_numeric(df['chemistry_score'], errors='coerce')
df['biology_score'] = pd.to_numeric(df['biology_score'], errors='coerce')
df['english_score'] = pd.to_numeric(df['english_score'], errors='coerce')
df['geography_score'] = pd.to_numeric(df['geography_score'], errors='coerce')
df
```
